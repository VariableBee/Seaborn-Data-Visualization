## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("diamonds")
#data.head()

plt.figure(figsize=(10, 6))
sns.set(style="darkgrid")
colors = sns.color_palette("pastel")

ax = sns.barplot(x='cut', y='price', data=data, palette=colors)

ax.set_xlabel('Tipo de Corte', fontsize=12)
ax.set_ylabel('Preço Médio', fontsize=12)
ax.set_title('Preço Médio do Diamante por Tipo de Corte', fontsize=14)

# Adiciona rótulos nas barras com os valores
for bar, c in zip(ax.patches, colors):
    height = bar.get_height()
    ax.text(
        x=bar.get_x() + bar.get_width() / 2,
        y=height + 50,
        s=f'${height:.2f}',
        ha='center',
        va='bottom',
        color=c,
        fontsize=10,
    )

sns.despine()

# Ajusta os espaçamentos dos subplots para evitar cortes de texto
plt.tight_layout()

# Adicionando sombras às barras
for bar in ax.patches:
    bar.set_edgecolor('black')  # Adiciona bordas pretas às barras
    bar.set_linewidth(1.5)  # Aumenta a largura das bordas

plt.show()
```

**Insight do gráfico 1**: Ainda que o preço do diamante é influenciado por outros fatores além do tipo de corte, como a quilatagem, a clareza, a cor e outros atributos. No gráfico de barras apresentado, podemos observar que os diamantes com corte "Ideal" e "Premium" têm preços médios mais altos em comparação com os diamantes de corte "Good", "Very Good" e "Fair". Isso sugere que o tipo de corte é um fator relevante na precificação dos diamantes, visto que ele influencia na forma que a luz é refletida pelo diamante, ou seja, em seu brilho.

No entanto, é importante ressaltar que o preço do diamante é influenciado por outros fatores, é essencial considerar todos esses fatores em conjunto ao avaliar a relação entre o preço do diamante e seu tipo de corte.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("flights")
#data.head()

plt.figure(figsize=(10, 6))
sns.set(style="darkgrid")
colors = sns.color_palette("pastel")

ax = sns.barplot(x='year', y='passengers', data=data, palette=colors)

ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Passageiros', fontsize=12)
ax.set_title('Número de Passageiros por Ano', fontsize=14)

sns.despine()

# Ajusta o tamanho da fonte dos rótulos dos eixos e do título
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title('Número de Passageiros por Ano', fontsize=14)

# Ajusta os espaçamentos dos subplots para evitar cortes de texto
plt.tight_layout()

# Adicionando sombras às barras
for bar in ax.patches:
    bar.set_edgecolor('black')  # Adiciona bordas pretas às barras
    bar.set_linewidth(1.5)  # Aumenta a largura das bordas

plt.show()
```

**Insight do gráfico 2**: O gráfico revela um aumento consistente no número de passageiros ao longo dos anos. Isso sugere um crescimento contínuo e sustentado da indústria da aviação, com uma demanda crescente por serviços de transporte aéreo.

Contexto da época: Na década de 1950, houve uma rápida evolução dos aviões a jato, com a introdução de aeronaves comerciais a jato como o Boeing 707 e o Douglas DC-8. Essas aeronaves eram mais rápidas, mais eficientes em termos de combustível e podiam transportar um maior número de passageiros em comparação com as aeronaves a hélice anteriores. A tecnologia a jato permitiu voos de longa distância com mais rapidez, tornando as viagens aéreas transcontinentais e intercontinentais mais acessíveis e populares.

Durante esse período, a aviação comercial também testemunhou um aumento no número de companhias aéreas, levando a uma expansão significativa da rede de rotas aéreas em todo o mundo. As cabines das aeronaves foram aprimoradas, oferecendo maior conforto aos passageiros, com assentos mais espaçosos e recursos adicionais, como refeições a bordo e entretenimento aéreo. A popularidade das viagens aéreas aumentou, e a aviação comercial se tornou uma opção preferida para viagens de longa distância.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("flights")

# Filtrar os dados no período de 1949 a 1959
filtered_data = data[(data['year'] >= 1949) & (data['year'] <= 1959)]

plt.figure(figsize=(10, 6))
sns.set(style="darkgrid")

ax = sns.lineplot(x='year', y='passengers', hue='month', data=filtered_data, palette='tab20', linewidth=2.5)

ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Passageiros', fontsize=12)
ax.set_title('Número de Passageiros por Mês (1949-1959)', fontsize=14)

sns.despine()

# Ajusta o tamanho da fonte dos rótulos dos eixos e do título
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.title('Número de Passageiros por Mês (1949-1959)', fontsize=14)

# Adiciona legenda fora do gráfico
handles, labels = ax.get_legend_handles_labels()
plt.legend(handles, labels, title='Mês', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

**Insight do gráfico 3**: A quantidade de passageiros apresenta um padrão geral de crescimento ao longo dos anos. O número de passageiros aumenta consideravelmente de ano para ano, indicando um aumento no uso da aviação comercial durante esse período como mencionado nos insight's anteriores.

Os meses de verão do hemisfério norte(junho, julho e agosto) apresentam picos de passageiros, indicando uma demanda maior durante as férias. É o período mais quente do ano em muitas regiões do hemisfério norte.

Os meses de dezembro e janeiro também registram um aumento significativo no número de passageiros, possivelmente devido às viagens de fim de ano e às festividades de Natal e Ano Novo.
